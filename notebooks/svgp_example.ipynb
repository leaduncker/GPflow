```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
N = 200
X = np.random.randn(N,1)
Y = X**3-X + np.random.randn(N,1)*1
plt.plot(X,Y,'x')
```

```python
import GPflow
k = GPflow.kernels.RBF(1)
lik = GPflow.likelihoods.Gaussian()
M=10
Z = np.random.randn(M,1)
m = GPflow.svgp.SVGP(X,Y,k,lik,Z)
```

```python
plt.plot(X,Y,'x')
plt.vlines(Z,-15,-10)
```

```python
r=m.optimize()
```

```python
mu,v=m.predict_f_full_cov(X)
m
```

```python
plt.plot(X,Y,'x')
plt.plot(X,mu,'o')
plt.vlines( m.Z.value,-15,-10)
```
